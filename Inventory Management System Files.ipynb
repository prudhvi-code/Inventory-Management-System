```python
import time

fd = open('inventorynew.txt','r')  
products = fd.read().split('\n')
fd.close()

username = input("Enter your Name: ")
userphone    = input("Enter your Phone No: ")
usermail     = input("Enter your Mail: ")
userprdid  = input("Enter product ID: ")
userquan  = input("Enter product Quantity: ")

updatedlst = []

for product in products:
    
    proddet = product.split(',')
    
    if(proddet[0] == userprdid):

        if (int(userquan) <= int(proddet[3])):
            print("-"*30)
            print("Product Name     : ", proddet[1])
            print("Price            : ", proddet[2]) 
            print("Quantity         : ", userquan) 
            print("-"*30)
            print("Billing Amount   : ", int(userquan) * int(proddet[2]))
            print("-"*30)

            proddet[3] = str(int(proddet[3]) - int(userquan))
 
            fd = open("sales.txt",'a')
            salesdet = username +","+ userphone +","+ usermail +","+proddet[1] +","+ userprdid +","+ userquan+","+ str(int(userquan) * int(proddet[2]))+","+time.ctime()+ "\n"
            fd.write(salesdet)
            fd.close()
            
        else:

            print("Sorry, We're not having enough Quantity.")
            print("We're having only",proddet[3],'quantity.')
            print("Would you like to purchase it?")
            
            ch = input("Press Y or N ")
            
            if (ch == 'Y' or ch == 'y'):

                print("-"*30)
                print("Product Name     : ", proddet[1])
                print("Price            : ", proddet[2]) 
                print("Quantity         : ", proddet[3]) 
                print("-"*30)
                print("Billing Amount   : ", int(proddet[3]) * int(proddet[2]))
                print("-"*30)

                proddet[3] = '0'

            else:
                print("Thank You Visit Again")

    updatedlst.append(proddet)
    

    
lst = []

for i in updatedlst:
    prod = i[0] +","+  i[1] +","+ i[2] +","+ i[3] + '\n'
    lst.append(prod)

    
lst[-1] = lst[-1][:-1]

fd = open('inventorynew.txt','w')

for i in lst:
    fd.write(i)

fd.close()

print("-"*30)
print("Inventory Updated")
```
